# Исследовательский анализ интернет-магазина "Стримчик"

## Описание проекта

Интернет-магазине «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


### Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.
Данные за 2016 год могут быть неполными.

### Смотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.

```python
# построим диаграмму рассеяния для ТОП-5 игровых платформ и вычислим корреляцию между 
# оценкой пользователей/критиков и общими продажами
for group_name, data in platform_top_5.groupby('platform'):
    fig, ax = plt.subplots(1, 2, figsize=(13,8))
    sns.scatterplot(x='user_score', y='total_sales', data=data, ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=data, ax=ax[1])
    fig.suptitle(group_name, fontsize=17)
    ax[0].set(xlabel='Оценка пользователей')
    ax[1].set(xlabel='Оценка критиков')
    ax[0].set(ylabel='Количество продаж')
    ax[1].set(ylabel='Количество продаж');
    plt.show()
    print(f'Корреляция между оценкой пользователей и общими продажами {group_name}:',\
          data['user_score'].corr(data['total_sales']))
    print(f'Корреляция между оценкой критиков и общими продажами {group_name}:',\
          data['critic_score'].corr(data['total_sales']))
    print('*'*100)
    
```

### Проверка гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем гипотезы следующим образом:  

H0 - средние пользовательские рейтинги платформ Xbox One и PC равны;  
Н1 - средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
# сделаем срез данных для платформы Xbox One
user_score_mean_xbox = data_current_period.query('platform == "XOne"').dropna(subset=['user_score'])
user_score_mean_xbox.head()
```

```python
# сделаем срез данных для семейства платформ PC (только для тех, которые попали в рейтинг ТОП-5)
user_score_mean_pc = data_current_period.query('platform == "PC"').dropna(subset=['user_score'])
user_score_mean_pc.head()
```

```python
# для проверки нулевой гипотезы средних пользовательских рейтингов независимых друг от друга данных применим
# т-тест ttest_ind с уровнем статистической значимости в 5%. Затем найдём p-value - вероятность найти 
# наблюдаемое значение на выборке или более отдалённое от предположенного в нулевой гипотезе
alpha = 0.05
results = st.ttest_ind(
    user_score_mean_xbox['user_score'], 
    user_score_mean_pc['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Предварительный вывод

Значение p-value в 14 % не оказалось меньше уровня статистической значимости, а значит можно предполагать, что средние пользовательские рейтинги платформ Xbox One и PC могут быть равны.

Статистические вывод:  
на имеющихся данных на уровнем значимости 5% (уровне доверия 95%) нет оснований отвергнуть нулевую гипотезу в пользу альтернативной.  

Содержательный вывод:  
средние пользовательские рейтинги платформ Xbox One и PC равны.

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем гипотезы следующим образом:  

H0 - средние пользовательские рейтинги жанров Action и Sports равны;  
Н1 - средние пользовательские рейтинги жанров Action и Sports не равны.

```python
# сделаем срез данных для жанра Action
user_score_action = data_current_period.query('genre == "Action"').dropna(subset=['user_score'])
user_score_action.head()
```

```python
# сделаем срез данных для жанра Sports
user_score_sports = data_current_period.query('genre == "Sports"').dropna(subset=['user_score'])
user_score_sports.head()
```

```python
# для проверки нулевой гипотезы средних пользовательских рейтингов независимых друг от друга данных применим
# т-тест ttest_ind с уровнем статистической значимости в 5%. Затем найдём p-value - вероятность найти 
# наблюдаемое значение на выборке или более отдалённое от предположенного в нулевой гипотезе
alpha = 0.05
results = st.ttest_ind(
    user_score_action['user_score'], 
    user_score_sports['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Предварительный вывод

Значение p-value оказалось очень маленьким и меньше уровня статистической значимости, а значит можно предполагать, что средние пользовательские рейтинги жанров Action и Sports не равны.

Статистические вывод:  
на имеющихся данных на уровнем значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативной.  

Содержательный вывод:  
средние пользовательские рейтинги жанров Action и Sports разные.
